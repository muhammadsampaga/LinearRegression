```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
datasets = pd.read_csv('Daftar_gaji.csv')

x = datasets.iloc[:, :-1].values
y = datasets.iloc[:, 1].values
```

```python
dataku = pd.DataFrame(datasets)

plt.scatter(dataku.Tahun_bekerja, dataku.Gaji)
plt.xlabel('tahun bekerja')
plt.ylabel('gaji')
plt.title('grafik masa kerja vs pendapatan')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regresi = LinearRegression()
regresi.fit(x_train, y_train)
```

```python
plt.figure(figsize=(5,4))

plt.scatter(x_train,y_train, color = 'blue')
plt.plot(x_train, regresi.predict(x_train), color = 'red')
plt.title('masa kerja vs pendapatan')
plt.xlabel('tahun bekerja')
plt.ylabel('gaji')
plt.show()
```

```python
plt.figure(figsize=(5,4))

plt.scatter(x_test,y_test, color = 'blue')
plt.plot(x_train, regresi.predict(x_train), color = 'red')
plt.title('masa kerja vs pendapatan testing')
plt.xlabel('tahun bekerja')
plt.ylabel('gaji')
plt.show()
```
